## splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25, train_size=0.75, random_state=1)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

## preprocessing steps on training data

```python
# seperating categorical and numerical columns
x_train.head()
```

```python
x_train.dtypes
```

```python
x_train_cat=x_train.select_dtypes(include=['object'])
x_train_cat.head()
```

```python
x_train['MultipleLines'].replace("No phone service","No",inplace=True)
x_train["OnlineSecurity"].replace("No internet service","No",inplace=True)
x_train["OnlineBackup"].replace("No internet service","No",inplace=True)
x_train["DeviceProtection"].replace("No internet service","No",inplace=True)
x_train["TechSupport"].replace("No internet service","No",inplace=True)
x_train["StreamingTV"].replace("No internet service","No",inplace=True)
x_train["StreamingMovies"].replace("No internet service","No",inplace=True)
```

```python
for i in x_train.select_dtypes(include=['object']):
    print(i)
    print(x_train[i].value_counts())
    print("\n")
```

```python
x_train_num=x_train.select_dtypes(include=['float64','int64'])
x_train_num.head()
```

## scaling the numerical features

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_num_rescaled=pd.DataFrame(scaler.fit_transform(x_train_num), 
                                    columns = x_train_num.columns, 
                                    index = x_train_num.index)
x_train_num_rescaled.head()
```

## applying one hot encoding on categorical columns

```python
from sklearn.preprocessing import OneHotEncoder
encoder_ = OneHotEncoder(drop='first',sparse=False)
x_train_cat_ohe=pd.DataFrame(encoder_.fit_transform(x_train_cat),
                             columns=encoder_.get_feature_names(x_train_cat.columns),
                              
                               index = x_train_cat.index)

x_train_cat_ohe.head()
```

```python
encoder_.categories_
```

```python
encoder_.get_params
```

```python
x_train_transformed=pd.concat([x_train_num_rescaled,x_train_cat_ohe],axis=1)
```

```python
x_train_transformed.head()
```

## applying same preprocessing steps on test data

```python
x_test['MultipleLines'].replace("No phone service","No",inplace=True)
x_test["OnlineSecurity"].replace("No internet service","No",inplace=True)
x_test["OnlineBackup"].replace("No internet service","No",inplace=True)
x_test["DeviceProtection"].replace("No internet service","No",inplace=True)
x_test["TechSupport"].replace("No internet service","No",inplace=True)
x_test["StreamingTV"].replace("No internet service","No",inplace=True)
x_test["StreamingMovies"].replace("No internet service","No",inplace=True)
```

## Training the model using knn algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
knn=KNeighborsClassifier(n_neighbors=22)
knn.fit(x_train_transformed,y_train)
y_pred=knn.predict(x_test_transformed)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

temp_df.head()
```

```python
metrics.accuracy_score(y_test,y_pred)
```

## Training the model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression(C=0.9,penalty='l2')
regressor.fit(x_train_transformed,y_train)
y_pred=regressor.predict(x_test_transformed)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

temp_df.head()
```

## Training the model using Support Vector Machines

## Training the model using Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier()
dtc=DTC.fit(x_train_transformed , y_train)
y_pred=dtc.predict(x_test_transformed)
metrics.accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier(min_samples_split=12,min_samples_leaf=10,max_features="sqrt",random_state=15)
dtc=DTC.fit(x_train_transformed , y_train)
y_pred=dtc.predict(x_test_transformed)
metrics.accuracy_score(y_test,y_pred)
```

## training the model with Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
RFC=rfc.fit(x_train_transformed,y_train)
y_pred=RFC.predict(x_test_transformed)
metrics.accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=1380,criterion='entropy',max_depth=60,bootstrap='True')
RFC=rfc.fit(x_train_transformed,y_train)
y_pred=RFC.predict(x_test_transformed)
metrics.accuracy_score(y_test,y_pred)
```
